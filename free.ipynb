```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'freelancer_earnings_bd.csv'
df = pd.read_csv(file_path)
```

```python
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
#Most Common Job Categories
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df['Job_Category'].value_counts().index,
            y=df['Job_Category'].value_counts().values)
plt.title('Most Common Job Categories')
plt.xticks(rotation=45)
plt.show()
```

```python
#The above Barplot shows that Graphic Design is the most common job in Freelancing.
```

```python
# Earnings Distribution
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['Earnings_USD'],bins=30, kde=True)
plt.title('Distribution Of Earnings')
plt.xlabel('Earnings in USD')
plt.show()
```

```python
# From the graph it can be observed, that a large number of individuals earn between $2000 to $4000 while 
#fewer individuals earn very high or very low incomes .
```

```python
# Experience Level vs Earnings
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Experience_Level', y='Earnings_USD', data=df)
plt.title('Earnings by Experience Level')
plt.show()
```

```python
# The plot shows the expert level individuals earning more and having a wider range of potential incomes.
```

```python
#Platform Comparison
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Platform', y='Earnings_USD', data=df,color='orange', estimator=sum)
plt.title('Total Earnings by Platform')
plt.show()
```

```python
# The graph shows that the Upwork Platform earns more as compared to the other platforms.
```

```python
# Client Rating Analysis
```

```python

df['Rating_Binned'] = pd.cut(df['Client_Rating'], bins=[0, 2, 3, 4, 5], labels=['0-2', '2-3', '3-4', '4-5'])

plt.figure(figsize=(10,6))
sns.countplot(x='Rating_Binned', data=df,color='darkblue')
plt.title('Distribution of Binned Client Ratings')
plt.xlabel('Client Rating Range')
plt.ylabel('Count')
plt.show()

```

```python
#The above graph shows that the client rate mostly between 3-4 the 4-5 and very rarely between 2-3.
```

```python
# Top Payment Methods
```

```python
payment_methods = df['Payment_Method'].value_counts()
plt.figure(figsize=(10,6))
plt.pie(payment_methods,labels=payment_methods.index,autopct='%1.1f%%')
plt.title('Top Payment Methods')
plt.show()
```

```python
# The Payments are mostly done in the form of Crypto , then Bank Transfer, then Mobile Banking and then PayPal
```

```python
# Job Success Rate
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Job_Category', y='Job_Success_Rate', data=df)
plt.title('Jobs Success Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
# The highest success rate is of the Graphic Designing Job.
```

```python
#Marketing Spent on Jobs
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Job_Category', y='Marketing_Spend',data=df,color='green')
plt.title('Marketing Spent on Each Job')
plt.xticks(rotation=45)
plt.show()
```

```python
# The above Graph shows that the most amount of marketing is done on the Data Entry jobs.
```

```python
# Return of Investment
#How many Dollars are earned per dollar of Marketing?
```

```python
df['ROI'] = df['Earnings_USD'] / df['Marketing_Spend']
print(df['ROI'])
```

```python
# Earnings Per Job
```

```python
df['Earnings_per_Job'] = df['Earnings_USD'] / df['Job_Completed']
print(df['Earnings_per_Job'])
```

```python
# Time effciency Per Job
```

```python
df['Days_per_Job'] = df['Job_Duration_Days'] / df['Job_Completed']
print(df['Days_per_Job'])
```

```python
# Hourly Value from All Work
```

```python
df['Total_Hours'] = df['Job_Duration_Days'] * 8  
df['Real_Hourly_Earnings'] = df['Earnings_USD'] / df['Total_Hours']
print(df['Real_Hourly_Earnings'])
```

```python
#Client Satisfaction Index
```

```python
df['Client_Satisfaction'] = (df['Client_Rating'] * 20 + df['Rehire_Rate']) / 2
print(df['Client_Satisfaction'])
```

```python
# Marketing Effeciency Score
```

```python
df['Marketing_Efficiency'] = df['Job_Completed'] / df['Marketing_Spend']
print(df['Marketing_Efficiency'])
```

```python
# Correlation Heatmap
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('Correlation Between Numerical Features')
plt.show()
```

```python
#The heatmap reveals that Earnings are most strongly driven by the number of jobs completed, 
# not just hourly rate or marketing spend. Additionally, 
# client satisfaction (ratings) is closely tied to rehire rate, 
# highlighting the value of building long-term client relationships over one-off gigs.
#  Interestingly, higher hourly rates may slightly correlate with lower success rates, 
# suggesting that premium pricing could come with higher expectations or tougher projects.
```
